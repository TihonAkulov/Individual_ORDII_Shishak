```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

# Загрузка и первичный обзор

```python
df = pd.read_csv("google_books_dataset.csv") 
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
df.shape
```

```python
df.dtypes
```

```python
df["date"] = pd.to_datetime(df["published_date"], format="%Y-%m-%d", errors="coerce")

df["day"] = df["date"].dt.day
df["month"] = df["date"].dt.month
df["year"] = df["date"].dt.year
```

# Обработка данных

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
print(df["authors"].value_counts())
print('----------------------------')
print(df["title"].value_counts())
print('----------------------------')
print(df["publisher"].value_counts())
print('----------------------------')
print(df["categories"].value_counts())
print('----------------------------')
print(df["language"].value_counts())
```

```python
df.nunique()
```

```python
corr = df.corr(numeric_only=True)
corr
```

# Визуализация данных

```python
plt.figure(figsize=(6,4))
plt.hist(df["page_count"], bins=50)
plt.title("Распределение кол-ва страниц")
plt.xlabel("Кол-во страниц")
plt.ylabel("Частота")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["date"], bins=50)
plt.title("Распределение дат публикации")
plt.xlabel("Дата публикации")
plt.ylabel("КОл-во книг")
plt.show()
```

```python
top_10_categories = df["categories"].value_counts().head(10)

plt.figure(figsize=(8, 4))
top_10_categories.plot(kind="bar")

plt.xticks(rotation=45, ha="right")
plt.title("Топ-10 категорий по количеству")
plt.show()
```

```python
top_10_categories = df["authors"].value_counts().head(10)

plt.figure(figsize=(8, 4))
top_10_categories.plot(kind="bar")

plt.xticks(rotation=45, ha="right")
plt.title("Топ-10 авторов по количеству")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Корреляционная матрица")
plt.show()
```
